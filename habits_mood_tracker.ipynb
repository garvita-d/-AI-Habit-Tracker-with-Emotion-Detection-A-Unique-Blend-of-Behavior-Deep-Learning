```python
# AI Habit Tracker with Enhanced Emotion Detection

import cv2
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import json
import os
from collections import defaultdict, Counter
import warnings
import time
warnings.filterwarnings('ignore')
from IPython.display import display, Javascript, clear_output
from google.colab.output import eval_js
from base64 import b64decode

try:
    from deepface import DeepFace
except ImportError:
    print("Installing DeepFace...")
    import subprocess
    import sys
    subprocess.check_call([sys.executable, "-m", "pip", "install", "deepface"])
    from deepface import DeepFace

class EmotionHabitTrackerColab:
    def __init__(self):
        self.emotion_log = []
        self.habits_log = []
        self.current_habit = None
        self.session_start = None
        self.emotion_counts = defaultdict(int)

        # Create data directory if it doesn't exist
        os.makedirs('habit_data', exist_ok=True)

        # Load existing data if available
        self.load_data()

    def load_data(self):
        """Load existing emotion and habit logs"""
        try:
            if os.path.exists('habit_data/emotion_log.csv'):
                df = pd.read_csv('habit_data/emotion_log.csv')
                self.emotion_log = df.to_dict('records')

            if os.path.exists('habit_data/habits_log.csv'):
                df = pd.read_csv('habit_data/habits_log.csv')
                self.habits_log = df.to_dict('records')

            print(f"Loaded {len(self.emotion_log)} emotion records and {len(self.habits_log)} habit sessions")
        except Exception as e:
            print(f"No existing data found or error loading: {e}")

    def save_data(self):
        """Save emotion and habit logs to CSV files"""
        try:
            # Save emotion log
            if self.emotion_log:
                df_emotions = pd.DataFrame(self.emotion_log)
                df_emotions.to_csv('habit_data/emotion_log.csv', index=False)

            # Save habits log
            if self.habits_log:
                df_habits = pd.DataFrame(self.habits_log)
                df_habits.to_csv('habit_data/habits_log.csv', index=False)

            print("Data saved successfully!")
        except Exception as e:
            print(f"Error saving data: {e}")

    def take_photo(self, filename='photo.jpg', quality=0.9):
        """Take photo using Google Colab webcam with improved settings"""
        js = Javascript('''
        async function takePhoto(quality) {
          const div = document.createElement('div');
          div.style.margin = '10px';

          const capture = document.createElement('button');
          capture.textContent = 'Capture Photo';
          capture.style.padding = '10px 20px';
          capture.style.fontSize = '16px';
          capture.style.backgroundColor = '#4CAF50';
          capture.style.color = 'white';
          capture.style.border = 'none';
          capture.style.borderRadius = '5px';
          capture.style.cursor = 'pointer';
          div.appendChild(capture);

          const video = document.createElement('video');
          video.style.display = 'block';
          video.style.width = '640px';
          video.style.height = '480px';
          video.style.margin = '10px 0';
          video.style.border = '2px solid #ddd';
          div.appendChild(video);

          const status = document.createElement('div');
          status.style.margin = '10px 0';
          status.style.fontSize = '14px';
          status.textContent = 'Starting camera... Please ensure good lighting and face the camera directly.';
          div.appendChild(status);

          document.body.appendChild(div);

          try {
            // Request higher quality video constraints
            const stream = await navigator.mediaDevices.getUserMedia({
              video: {
                width: { ideal: 640 },
                height: { ideal: 480 },
                facingMode: 'user',
                frameRate: { ideal: 30 }
              }
            });

            video.srcObject = stream;
            await video.play();
            status.textContent = 'Camera ready! Ensure good lighting and look directly at camera. Click "Capture Photo" when ready.';

            google.colab.output.setIframeHeight(document.documentElement.scrollHeight, true);

            await new Promise((resolve) => capture.onclick = resolve);

            // Wait a moment for better capture
            await new Promise(resolve => setTimeout(resolve, 100));

            const canvas = document.createElement('canvas');
            canvas.width = video.videoWidth;
            canvas.height = video.videoHeight;
            const ctx = canvas.getContext('2d');
            ctx.drawImage(video, 0, 0);

            stream.getTracks().forEach(track => track.stop());
            div.remove();

            return canvas.toDataURL('image/jpeg', quality);
          } catch (error) {
            status.textContent = 'Error accessing camera: ' + error.message;
            console.error('Camera error:', error);
            return null;
          }
        }
        ''')
        display(js)

        try:
            data = eval_js('takePhoto({})'.format(quality))
            if data:
                binary = b64decode(data.split(',')[1])
                with open(filename, 'wb') as f:
                    f.write(binary)

                img = cv2.imdecode(np.frombuffer(binary, np.uint8), cv2.IMREAD_COLOR)
                return img
            else:
                print("Failed to capture photo")
                return None
        except Exception as e:
            print(f"Error capturing photo: {e}")
            return None

    def preprocess_image(self, img):
        """Preprocess image for better emotion detection"""
        try:
            # Convert to RGB if needed
            if len(img.shape) == 3 and img.shape[2] == 3:
                img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
            else:
                img_rgb = img

            # Enhance contrast and brightness
            lab = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2LAB)
            l, a, b = cv2.split(lab)

            # Apply CLAHE (Contrast Limited Adaptive Histogram Equalization)
            clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
            l = clahe.apply(l)

            enhanced = cv2.merge([l, a, b])
            enhanced_rgb = cv2.cvtColor(enhanced, cv2.COLOR_LAB2RGB)

            return enhanced_rgb
        except Exception as e:
            print(f"Error preprocessing image: {e}")
            return img

    def convert_numpy_types(self, obj):
        """Convert numpy types to native Python types for JSON serialization"""
        if isinstance(obj, np.integer):
            return int(obj)
        elif isinstance(obj, np.floating):
            return float(obj)
        elif isinstance(obj, np.ndarray):
            return obj.tolist()
        elif isinstance(obj, dict):
            return {key: self.convert_numpy_types(value) for key, value in obj.items()}
        elif isinstance(obj, list):
            return [self.convert_numpy_types(item) for item in obj]
        else:
            return obj

    def analyze_emotion_from_image(self, img):
        """Analyze emotion from a single image with improved accuracy"""
        try:
            # Preprocess image for better detection
            processed_img = self.preprocess_image(img)

            # Try multiple models and backends for better accuracy
            models = ['VGG-Face', 'Facenet', 'OpenFace']
            backends = ['opencv', 'ssd', 'mtcnn']

            best_result = None
            highest_confidence = 0

            for model in models:
                for backend in backends:
                    try:
                        result = DeepFace.analyze(
                            processed_img,
                            actions=['emotion'],
                            enforce_detection=False,
                            detector_backend=backend,
                            model_name=model,
                            silent=True
                        )

                        if isinstance(result, list):
                            result = result[0]

                        # Get the confidence of the dominant emotion
                        dominant_emotion = result['dominant_emotion']
                        confidence = result['emotion'][dominant_emotion]

                        # Keep the result with highest confidence
                        if confidence > highest_confidence:
                            highest_confidence = confidence
                            best_result = result

                    except Exception:
                        continue  # Try next combination

            # Fallback to default if no good result
            if best_result is None:
                result = DeepFace.analyze(
                    processed_img,
                    actions=['emotion'],
                    enforce_detection=False,
                    silent=True
                )
                if isinstance(result, list):
                    result = result[0]
                best_result = result

            dominant_emotion = best_result['dominant_emotion']
            emotion_scores = best_result['emotion']

            # Apply confidence threshold and emotion refinement
            refined_emotion, refined_confidence = self.refine_emotion_detection(emotion_scores)

            # Convert numpy types to native Python types
            emotion_scores = self.convert_numpy_types(emotion_scores)

            return refined_emotion, emotion_scores, refined_confidence

        except Exception as e:
            print(f"Error analyzing emotion: {e}")
            return None, None, None

    def refine_emotion_detection(self, emotion_scores):
        """Refine emotion detection based on score differences and thresholds"""
        # Convert to regular dict if needed
        if hasattr(emotion_scores, 'items'):
            scores = dict(emotion_scores)
        else:
            scores = emotion_scores

        # Sort emotions by score
        sorted_emotions = sorted(scores.items(), key=lambda x: x[1], reverse=True)

        # Get top 2 emotions
        first_emotion, first_score = sorted_emotions[0]
        second_emotion, second_score = sorted_emotions[1] if len(sorted_emotions) > 1 else (None, 0)

        # Calculate confidence difference
        confidence_diff = first_score - second_score

        # If confidence difference is too small, might be uncertain
        if confidence_diff < 10:  # Less than 10% difference
            # Check if neutral is in top 2, might indicate unclear expression
            if 'neutral' in [first_emotion, second_emotion]:
                return 'neutral', first_score
            else:
                # Return the higher scoring emotion but with adjusted confidence
                return first_emotion, first_score * 0.8  # Reduce confidence due to uncertainty

        # Apply minimum confidence threshold
        if first_score < 30:  # Less than 30% confidence
            return 'uncertain', first_score

        # Special handling for similar emotions
        emotion_groups = {
            'negative': ['sad', 'angry', 'fear', 'disgust'],
            'positive': ['happy', 'surprise'],
            'calm': ['neutral']
        }

        # Check if top emotions are from same group
        for group_name, group_emotions in emotion_groups.items():
            if first_emotion in group_emotions and second_emotion in group_emotions:
                # Both emotions in same group, return the stronger one
                return first_emotion, first_score

        return first_emotion, first_score

    def display_emotion_analysis(self, img, emotion, emotion_scores, confidence):
        """Display comprehensive emotion analysis results"""
        print(f"\n DETAILED EMOTION ANALYSIS:")
        print(f"{'='*50}")
        print(f"Dominant Emotion: {emotion.upper()}")
        print(f"Confidence: {confidence:.1f}%")

        # Show all emotion scores
        print(f"\n📈 All Emotion Scores:")
        sorted_emotions = sorted(emotion_scores.items(), key=lambda x: x[1], reverse=True)
        for i, (emo, score) in enumerate(sorted_emotions, 1):
            bar = '█' * int(score / 5)  # Visual bar representation
            status = "🥇" if i == 1 else "🥈" if i == 2 else "🥉" if i == 3 else "  "
            print(f"{status} {emo.capitalize():10} {score:5.1f}% {bar}")

        # Interpretation
        print(f"\n Interpretation:")
        if confidence > 70:
            print(f"   High confidence - Clear {emotion} expression detected")
        elif confidence > 50:
            print(f"   Moderate confidence - Likely {emotion} with some ambiguity")
        elif confidence > 30:
            print(f"   Low confidence - Possible {emotion}, expression unclear")
        else:
            print(f"   Very low confidence - Expression difficult to interpret")

        # Show difference between top emotions
        if len(sorted_emotions) >= 2:
            diff = sorted_emotions[0][1] - sorted_emotions[1][1]
            print(f"   Confidence gap: {diff:.1f}% between top emotions")
            if diff < 10:
                print(f" Small gap suggests mixed or subtle expression")

        # Display the image with enhanced info
        plt.figure(figsize=(12, 6))

        # Original image
        plt.subplot(1, 2, 1)
        plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
        plt.title(f'Captured Image\nEmotion: {emotion.upper()} ({confidence:.1f}%)', fontsize=12)
        plt.axis('off')

        # Emotion score chart
        plt.subplot(1, 2, 2)
        emotions = [emo for emo, score in sorted_emotions]
        scores = [score for emo, score in sorted_emotions]
        colors = plt.cm.viridis(np.linspace(0, 1, len(emotions)))

        bars = plt.barh(emotions, scores, color=colors)
        plt.xlabel('Confidence (%)')
        plt.title('Emotion Score Breakdown')
        plt.xlim(0, 100)

        # Add value labels on bars
        for bar, score in zip(bars, scores):
            plt.text(score + 1, bar.get_y() + bar.get_height()/2,
                    f'{score:.1f}%', va='center', fontsize=9)

        plt.tight_layout()
        plt.show()

    def start_habit_session(self, habit_name):
        """Start tracking a specific habit"""
        self.current_habit = habit_name
        self.session_start = datetime.now()
        self.emotion_counts = defaultdict(int)
        print(f"\n Started tracking: {habit_name}")
        print(f"Session started at: {self.session_start.strftime('%H:%M:%S')}")
        print(f" Tip: Take photos at different points during your session for better emotion tracking!")

    def end_habit_session(self):
        """End current habit session and log summary"""
        if self.current_habit and self.session_start:
            session_end = datetime.now()
            duration = (session_end - self.session_start).total_seconds() / 60  # in minutes

            # Calculate dominant emotion for this session
            if self.emotion_counts:
                dominant_emotion = max(self.emotion_counts, key=self.emotion_counts.get)
                emotion_distribution = dict(self.emotion_counts)
            else:
                dominant_emotion = "unknown"
                emotion_distribution = {}

            # Log the habit session
            habit_entry = {
                'habit': self.current_habit,
                'start_time': self.session_start.strftime("%Y-%m-%d %H:%M:%S"),
                'end_time': session_end.strftime("%Y-%m-%d %H:%M:%S"),
                'duration_minutes': round(duration, 2),
                'dominant_emotion': dominant_emotion,
                'emotion_distribution': json.dumps(emotion_distribution),
                'total_emotion_samples': sum(self.emotion_counts.values())
            }

            self.habits_log.append(habit_entry)

            print(f"\n✅ Session completed!")
            print(f"Habit: {self.current_habit}")
            print(f"Duration: {duration:.1f} minutes")
            print(f"Dominant emotion: {dominant_emotion}")
            print(f"Emotion samples: {sum(self.emotion_counts.values())}")

            # Show emotion breakdown for session
            if emotion_distribution:
                print(f"\nSession emotion breakdown:")
                for emotion, count in sorted(emotion_distribution.items(), key=lambda x: x[1], reverse=True):
                    percentage = (count / sum(emotion_distribution.values())) * 100
                    print(f"  {emotion}: {count} samples ({percentage:.1f}%)")

            # Reset session
            self.current_habit = None
            self.session_start = None
            self.emotion_counts = defaultdict(int)

    def capture_and_analyze_emotion(self):
        """Capture photo and analyze emotion with enhanced feedback"""
        print("Preparing to capture photo for emotion analysis...")
        print("Tips for better detection:")
        print(" Ensure good lighting (face camera directly)")
        print(" Keep a natural, clear facial expression")
        print(" Remove glasses/masks if possible")
        print(" Stay still during capture")

        img = self.take_photo()
        if img is None:
            print(" Failed to capture photo")
            return

        print("Analyzing emotion (this may take a moment)...")
        result = self.analyze_emotion_from_image(img)

        if len(result) == 3 and result[0] is not None:
            dominant_emotion, emotion_scores, confidence = result
        else:
            print(" Could not analyze emotion from the image")
            print(" Try again with better lighting or clearer facial expression")
            return

        # Log emotion data
        timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")

        # Ensure all emotion scores are JSON serializable
        serializable_scores = self.convert_numpy_types(emotion_scores)

        emotion_entry = {
            'timestamp': timestamp,
            'emotion': dominant_emotion,
            'confidence': round(float(confidence), 2),
            'habit': self.current_habit if self.current_habit else 'none',
            'all_scores': json.dumps({k: round(float(v), 2) for k, v in serializable_scores.items()})
        }

        self.emotion_log.append(emotion_entry)

        # Update session emotion counts
        if self.current_habit:
            self.emotion_counts[dominant_emotion] += 1

        # Display comprehensive results
        self.display_emotion_analysis(img, dominant_emotion, serializable_scores, confidence)

        if self.current_habit:
            print(f"\n Current Session Info:")
            print(f"   Habit: {self.current_habit}")
            duration = (datetime.now() - self.session_start).total_seconds() / 60
            print(f"   Duration: {duration:.1f} minutes")
            print(f"   Emotion samples: {sum(self.emotion_counts.values())}")

    def interactive_session_mode(self):
        """Interactive session mode with better guidance"""
        print(" INTERACTIVE HABIT TRACKING SESSION")
        print("=" * 50)

        # Select habit
        habits = [
            "Study/Work", "Meditation", "Exercise", "Reading",
            "Coding", "Creative Work", "Music", "Cleaning"
        ]

        print("Available habits:")
        for i, habit in enumerate(habits, 1):
            print(f"{i}. {habit}")

        try:
            choice = int(input("Select habit (1-8) or 0 for custom: "))
            if choice == 0:
                habit_name = input("Enter custom habit name: ").strip()
            elif 1 <= choice <= 8:
                habit_name = habits[choice - 1]
            else:
                print("Invalid choice")
                return
        except ValueError:
            print("Invalid input")
            return

        # Start session
        self.start_habit_session(habit_name)

        print(f"\n Session Instructions:")
        print(f"• Take photos at different points during your {habit_name} session")
        print(f"• Each photo will be analyzed for emotions with detailed feedback")
        print(f"• Type 'photo' to capture and analyze your current emotion")
        print(f"• Type 'status' to see current session info")
        print(f"• Type 'done' when you finish the session")

        while True:
            command = input(f"\n[{self.current_habit}] Command (photo/status/done): ").strip().lower()

            if command == 'photo':
                self.capture_and_analyze_emotion()
            elif command == 'status':
                if self.current_habit and self.session_start:
                    duration = (datetime.now() - self.session_start).total_seconds() / 60
                    print(f"\n📊 Current Session Status:")
                    print(f"   Habit: {self.current_habit}")
                    print(f"   Duration: {duration:.1f} minutes")
                    print(f"   Emotion samples: {sum(self.emotion_counts.values())}")
                    if self.emotion_counts:
                        print(f"   Emotions captured:")
                        for emotion, count in sorted(self.emotion_counts.items(), key=lambda x: x[1], reverse=True):
                            print(f"     {emotion}: {count}")
            elif command == 'done':
                self.end_habit_session()
                break
            else:
                print("Invalid command. Use 'photo', 'status', or 'done'")

    def demo_mode_without_camera(self):
        """Demo mode that simulates emotion detection without camera"""
        print("\n DEMO MODE - Simulating emotion detection")
        print("This will generate sample data for testing the analysis features")

        habits = ["Study/Work", "Meditation", "Exercise", "Reading", "Coding"]
        emotions = ["happy", "neutral", "focused", "calm", "excited", "tired"]

        # Generate sample data
        for _ in range(50):  # Generate 50 sample emotion records
            habit = np.random.choice(habits) if np.random.random() > 0.3 else 'none'
            emotion = np.random.choice(emotions)
            timestamp = datetime.now() - timedelta(minutes=np.random.randint(0, 1440))  # Random time in last 24h

            emotion_entry = {
                'timestamp': timestamp.strftime("%Y-%m-%d %H:%M:%S"),
                'emotion': emotion,
                'confidence': round(np.random.uniform(70, 95), 2),
                'habit': habit,
                'all_scores': json.dumps({e: round(np.random.uniform(5, 25), 2) if e != emotion
                                        else round(np.random.uniform(70, 95), 2) for e in emotions})
            }
            self.emotion_log.append(emotion_entry)

        # Generate sample habit sessions
        for habit in habits:
            for _ in range(np.random.randint(2, 6)):  # 2-5 sessions per habit
                duration = round(np.random.uniform(15, 120), 2)  # 15-120 minutes
                dominant_emotion = np.random.choice(emotions)
                start_time = datetime.now() - timedelta(hours=np.random.randint(1, 48))

                habit_entry = {
                    'habit': habit,
                    'start_time': start_time.strftime("%Y-%m-%d %H:%M:%S"),
                    'end_time': (start_time + timedelta(minutes=duration)).strftime("%Y-%m-%d %H:%M:%S"),
                    'duration_minutes': duration,
                    'dominant_emotion': dominant_emotion,
                    'emotion_distribution': json.dumps({emotion: np.random.randint(1, 10) for emotion in emotions}),
                    'total_emotion_samples': np.random.randint(5, 20)
                }
                self.habits_log.append(habit_entry)

        self.save_data()
        print(f" Generated {len(self.emotion_log)} emotion records and {len(self.habits_log)} habit sessions")

    def analyze_patterns(self):
        """Analyze emotion patterns and generate insights"""
        if not self.emotion_log:
            print("No emotion data available for analysis")
            return

        df_emotions = pd.DataFrame(self.emotion_log)
        df_emotions['timestamp'] = pd.to_datetime(df_emotions['timestamp'])
        df_emotions['hour'] = df_emotions['timestamp'].dt.hour
        df_emotions['day_of_week'] = df_emotions['timestamp'].dt.day_name()

        print("📊 EMOTION ANALYSIS REPORT")
        print("=" * 50)

        # Overall emotion distribution
        emotion_counts = df_emotions['emotion'].value_counts()
        print(f"\n🎭 Overall Emotion Distribution:")
        for emotion, count in emotion_counts.items():
            percentage = (count / len(df_emotions)) * 100
            print(f"  {emotion.capitalize()}: {count} ({percentage:.1f}%)")

        # Emotions by habit
        if self.habits_log:
            df_habits = pd.DataFrame(self.habits_log)
            print(f"\n🎯 Habit Sessions Summary:")
            print(f"  Total sessions: {len(df_habits)}")
            print(f"  Total time tracked: {df_habits['duration_minutes'].sum():.1f} minutes")

            habit_emotions = df_habits.groupby('habit')['dominant_emotion'].apply(list)
            for habit, emotions in habit_emotions.items():
                emotion_counter = Counter(emotions)
                most_common = emotion_counter.most_common(1)[0]
                print(f"  {habit}: Most common emotion = {most_common[0]} ({most_common[1]} sessions)")

        # Time-based patterns
        hourly_emotions = df_emotions.groupby('hour')['emotion'].apply(lambda x: x.mode().iloc[0] if not x.empty else 'unknown')
        print(f"\n⏰ Peak Emotion by Hour of Day:")
        for hour, emotion in hourly_emotions.items():
            print(f"  {hour:02d}:00 - {emotion.capitalize()}")

    def create_visualizations(self):
        """Create comprehensive visualizations of emotion and habit data"""
        if not self.emotion_log:
            print("No data available for visualization")
            return

        df_emotions = pd.DataFrame(self.emotion_log)
        df_emotions['timestamp'] = pd.to_datetime(df_emotions['timestamp'])
        df_emotions['hour'] = df_emotions['timestamp'].dt.hour
        df_emotions['day_of_week'] = df_emotions['timestamp'].dt.day_name()
        df_emotions['date'] = df_emotions['timestamp'].dt.date

        # Create subplots
        fig = make_subplots(
            rows=2, cols=2,
            subplot_titles=('Emotion Distribution', 'Emotions Over Time',
                          'Emotions by Hour', 'Daily Emotion Trends'),
            specs=[[{"type": "pie"}, {"type": "scatter"}],
                  [{"type": "bar"}, {"type": "scatter"}]]
        )

        # 1. Emotion Distribution Pie Chart
        emotion_counts = df_emotions['emotion'].value_counts()
        fig.add_trace(
            go.Pie(labels=emotion_counts.index, values=emotion_counts.values, name="Emotions"),
            row=1, col=1
        )

        # 2. Emotions Over Time
        emotion_colors = {
            'happy': 'gold', 'sad': 'blue', 'angry': 'red', 'fear': 'purple',
            'surprise': 'orange', 'disgust': 'green', 'neutral': 'gray',
            'focused': 'darkblue', 'calm': 'lightblue', 'excited': 'red', 'tired': 'brown'
        }

        for emotion in df_emotions['emotion'].unique():
            emotion_data = df_emotions[df_emotions['emotion'] == emotion]
            fig.add_trace(
                go.Scatter(x=emotion_data['timestamp'], y=[emotion]*len(emotion_data),
                         mode='markers', name=emotion,
                         marker=dict(color=emotion_colors.get(emotion, 'black'), size=8)),
                row=1, col=2
            )

        # 3. Emotions by Hour
        hourly_counts = df_emotions.groupby(['hour', 'emotion']).size().unstack(fill_value=0)
        for emotion in hourly_counts.columns:
            fig.add_trace(
                go.Bar(x=hourly_counts.index, y=hourly_counts[emotion],
                      name=f"{emotion} (hourly)",
                      marker_color=emotion_colors.get(emotion, 'black')),
                row=2, col=1
            )

        # 4. Daily Emotion Trends
        daily_emotions = df_emotions.groupby(['date', 'emotion']).size().unstack(fill_value=0)
        for emotion in daily_emotions.columns:
            fig.add_trace(
                go.Scatter(x=daily_emotions.index, y=daily_emotions[emotion],
                          mode='lines+markers', name=f"{emotion} (daily)",
                          line=dict(color=emotion_colors.get(emotion, 'black'))),
                row=2, col=2
            )

        fig.update_layout(height=800, showlegend=True, title_text="AI Habit Tracker - Emotion Analysis Dashboard")
        fig.show()

        # Create habit-specific analysis if available
        if self.habits_log:
            self.create_habit_visualizations()

    def create_habit_visualizations(self):
        """Create habit-specific visualizations"""
        df_habits = pd.DataFrame(self.habits_log)
        df_habits['start_time'] = pd.to_datetime(df_habits['start_time'])
        df_habits['end_time'] = pd.to_datetime(df_habits['end_time'])
        df_habits['date'] = df_habits['start_time'].dt.date

        # Create habit analysis subplots
        fig = make_subplots(
            rows=2, cols=2,
            subplot_titles=('Habit Session Duration', 'Habit Frequency',
                          'Emotions by Habit', 'Habit Performance Over Time'),
            specs=[[{"type": "bar"}, {"type": "bar"}],
                  [{"type": "bar"}, {"type": "scatter"}]]
        )

        # 1. Average session duration by habit
        avg_duration = df_habits.groupby('habit')['duration_minutes'].mean()
        fig.add_trace(
            go.Bar(x=avg_duration.index, y=avg_duration.values,
                  name="Avg Duration (min)", marker_color='lightblue'),
            row=1, col=1
        )

        # 2. Habit frequency
        habit_counts = df_habits['habit'].value_counts()
        fig.add_trace(
            go.Bar(x=habit_counts.index, y=habit_counts.values,
                  name="Session Count", marker_color='lightgreen'),
            row=1, col=2
        )

        # 3. Dominant emotions by habit
        habit_emotion_counts = df_habits.groupby(['habit', 'dominant_emotion']).size().unstack(fill_value=0)
        for emotion in habit_emotion_counts.columns:
            fig.add_trace(
                go.Bar(x=habit_emotion_counts.index, y=habit_emotion_counts[emotion],
                      name=f"{emotion}", offsetgroup=emotion),
                row=2, col=1
            )

        # 4. Habit performance over time (total minutes per day)
        daily_habits = df_habits.groupby(['date', 'habit'])['duration_minutes'].sum().unstack(fill_value=0)
        for habit in daily_habits.columns:
            fig.add_trace(
                go.Scatter(x=daily_habits.index, y=daily_habits[habit],
                          mode='lines+markers', name=f"{habit} (daily mins)"),
                row=2, col=2
            )

        fig.update_layout(height=800, showlegend=True,
                         title_text="Habit Analysis Dashboard")
        fig.show()

        # Additional insights
        print(f"\n HABIT INSIGHTS:")
        print(f"   Most frequent habit: {habit_counts.index[0]} ({habit_counts.iloc[0]} sessions)")
        print(f"   Longest average sessions: {avg_duration.index[-1]} ({avg_duration.iloc[-1]:.1f} min)")
        print(f"   Total tracking time: {df_habits['duration_minutes'].sum():.1f} minutes")
        print(f"   Average session length: {df_habits['duration_minutes'].mean():.1f} minutes")

    def export_data(self, format='csv'):
        """Export emotion and habit data"""
        try:
            timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")

            if format.lower() == 'csv':
                # Export emotions
                if self.emotion_log:
                    df_emotions = pd.DataFrame(self.emotion_log)
                    filename = f'emotion_export_{timestamp}.csv'
                    df_emotions.to_csv(filename, index=False)
                    print(f" Emotions exported to {filename}")

                # Export habits
                if self.habits_log:
                    df_habits = pd.DataFrame(self.habits_log)
                    filename = f'habits_export_{timestamp}.csv'
                    df_habits.to_csv(filename, index=False)
                    print(f" Habits exported to {filename}")

            elif format.lower() == 'json':
                # Export as JSON
                export_data = {
                    'emotions': self.emotion_log,
                    'habits': self.habits_log,
                    'export_timestamp': datetime.now().isoformat(),
                    'total_emotion_records': len(self.emotion_log),
                    'total_habit_sessions': len(self.habits_log)
                }

                filename = f'habit_tracker_export_{timestamp}.json'
                with open(filename, 'w') as f:
                    json.dump(export_data, f, indent=2)
                print(f" Data exported to {filename}")

        except Exception as e:
            print(f" Error exporting data: {e}")

    def get_insights_and_recommendations(self):
        """Generate personalized insights and recommendations"""
        if not self.emotion_log:
            print("No data available for insights")
            return

        df_emotions = pd.DataFrame(self.emotion_log)
        df_emotions['timestamp'] = pd.to_datetime(df_emotions['timestamp'])
        df_emotions['hour'] = df_emotions['timestamp'].dt.hour

        print(" PERSONALIZED INSIGHTS & RECOMMENDATIONS")
        print("=" * 60)

        # Emotion patterns
        most_common_emotion = df_emotions['emotion'].mode().iloc[0]
        avg_confidence = df_emotions['confidence'].mean()

        print(f"\n Your Emotional Profile:")
        print(f"   Most common emotion: {most_common_emotion.capitalize()}")
        print(f"   Average detection confidence: {avg_confidence:.1f}%")

        # Time-based insights
        morning_emotions = df_emotions[df_emotions['hour'] < 12]['emotion'].mode()
        evening_emotions = df_emotions[df_emotions['hour'] >= 18]['emotion'].mode()

        if not morning_emotions.empty and not evening_emotions.empty:
            print(f"   Morning emotion: {morning_emotions.iloc[0].capitalize()}")
            print(f"   Evening emotion: {evening_emotions.iloc[0].capitalize()}")

        # Habit-based insights
        if self.habits_log:
            df_habits = pd.DataFrame(self.habits_log)

            # Best performing habits (positive emotions)
            positive_emotions = ['happy', 'calm', 'excited', 'focused']
            habit_positivity = {}

            for habit in df_habits['habit'].unique():
                habit_emotions = df_habits[df_habits['habit'] == habit]['dominant_emotion'].tolist()
                positive_count = sum(1 for emotion in habit_emotions if emotion in positive_emotions)
                positivity_score = positive_count / len(habit_emotions) if habit_emotions else 0
                habit_positivity[habit] = positivity_score

            best_habit = max(habit_positivity, key=habit_positivity.get) if habit_positivity else None
            worst_habit = min(habit_positivity, key=habit_positivity.get) if habit_positivity else None

            print(f"\n Habit Performance:")
            if best_habit:
                print(f"   Most positive habit: {best_habit} ({habit_positivity[best_habit]*100:.1f}% positive emotions)")
            if worst_habit:
                print(f"   Needs attention: {worst_habit} ({habit_positivity[worst_habit]*100:.1f}% positive emotions)")

        # Recommendations
        print(f"\n Personalized Recommendations:")

        if most_common_emotion in ['sad', 'tired', 'angry']:
            print(f"Consider adding meditation or relaxation sessions to improve mood")
            print(f"Physical exercise might help boost positive emotions")

        if most_common_emotion == 'neutral':
            print(f" Try engaging in more stimulating or creative activities")
            print(f" Social activities might help increase positive emotions")

        if avg_confidence < 70:
            print(f" For better emotion detection, ensure good lighting during photo capture")
            print(f" Try more expressive facial expressions for clearer readings")

        # Time-based recommendations
        low_energy_hours = df_emotions.groupby('hour')['emotion'].apply(
            lambda x: (x == 'tired').sum() / len(x) if len(x) > 0 else 0
        ).sort_values(ascending=False).head(2)

        if not low_energy_hours.empty and low_energy_hours.iloc[0] > 0.3:
            print(f" You tend to feel tired around {low_energy_hours.index[0]}:00")
            print(f" Consider scheduling breaks or energy-boosting activities at this time")

    def main_menu(self):
        """Main interactive menu for the application"""
        print("\n AI HABIT TRACKER with Enhanced Emotion Detection")
        print("=" * 60)
        print("Welcome! This app tracks your emotions during different activities.")

        while True:
            print(f"\n MAIN MENU:")
            print("1.Start Interactive Session (with camera)")
            print("2.Quick Emotion Check")
            print("3.View Analytics & Patterns")
            print("4.Create Visualizations")
            print("5.Get Insights & Recommendations")
            print("6.Save Data")
            print("7.Export Data")
            print("8.Demo Mode (without camera)")
            print("9.Exit")

            try:
                choice = input("\nSelect option (1-9): ").strip()

                if choice == '1':
                    self.interactive_session_mode()
                elif choice == '2':
                    self.capture_and_analyze_emotion()
                elif choice == '3':
                    self.analyze_patterns()
                elif choice == '4':
                    self.create_visualizations()
                elif choice == '5':
                    self.get_insights_and_recommendations()
                elif choice == '6':
                    self.save_data()
                elif choice == '7':
                    format_choice = input("Export format (csv/json): ").strip().lower()
                    self.export_data(format_choice if format_choice in ['csv', 'json'] else 'csv')
                elif choice == '8':
                    self.demo_mode_without_camera()
                elif choice == '9':
                    print("Saving data before exit...")
                    self.save_data()
                    print("Thank you for using AI Habit Tracker!")
                    break
                else:
                    print("Invalid choice. Please select 1-9.")

            except KeyboardInterrupt:
                print("\n Saving data before exit...")
                self.save_data()
                print("Goodbye!")
                break
            except Exception as e:
                print(f"An error occurred: {e}")
                print("Please try again.")

if __name__ == "__main__":
    # Create and run the application
    try:
        app = EmotionHabitTrackerColab()

        # Check if running in Colab
        try:
            import google.colab
            print("Google Colab environment detected")
            print("Camera features will be available")
            app.main_menu()
        except ImportError:
            print("Running outside Google Colab")
            print("Camera features may not work - use demo mode")

            # Offer demo mode for non-Colab environments
            demo_choice = input("Run demo mode with sample data? (y/n): ").strip().lower()
            if demo_choice == 'y':
                app.demo_mode_without_camera()
                app.main_menu()
            else:
                print("Note: Some features require Google Colab environment")
                app.main_menu()

    except Exception as e:
        print(f"Error starting application: {e}")
        print("Please ensure all required packages are installed")

    print("\n Application Features Summary:")
    print("Real-time emotion detection using DeepFace")
    print("Habit tracking with emotion correlation")
    print("Interactive photo capture (Google Colab)")
    print("Advanced emotion analysis with confidence scoring")
    print("Comprehensive data visualization")
    print("Personalized insights and recommendations")
    print("Data export in CSV and JSON formats")
    print("Session-based habit tracking")
    print("Time-based emotion pattern analysis")
```
